LINEAR REGRESSION

```python
#    
from sklearn.linear_model import LinearRegression
#           
from random import randint
#       
import matplotlib.pyplot as plt
#      
import numpy as np
#       
linearity=LinearRegression()
#     20 
#    
train_x=[]
train_y=[]
for i in range(20):
   train_x.append([randint(0,20)])
   train_y.append([randint(0,20)])
#       
plt.plot(train_x,train_y,'ro')
#       
model=linearity.fit(train_x,train_y)
#     
r_sq = model.score(train_x,train_y)
x = np. linspace(0,20,100)
y = float(model.coef_) * x + float(model.intercept_)
plt. plot(x, y, '-r', label='my linear line')
print('coefficient of determination:', r_sq)

#   
print('intercept:', model.intercept_)

#        
print('slope:', model.coef_) 

#       ( )
y_pred = model.predict(train_x)
print('Predicted response:', y_pred, sep='\n')


```

Logistic Regression

```python
##    
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import math

###  
x_train=[]
y_train=[]
##  
for i in range(-10,10,1):
     x_train.append([i])
     y_train.append(randint(0,1))
 ##     
logistic=LogisticRegression(solver='liblinear', C=10.0, random_state=0)
##  
logistic.fit(x_train,y_train)
##  
x=np.linspace(-20,20,100)
y=1/(1+np.exp(-x))
##   
plt.ylim(top=1)  
plt.ylim(bottom=0)
##  
x_linear=np.linspace(-20,20,100)
y_linear=float(logistic.coef_)*x+float(logistic.intercept_)
##  Y=1/2
plt.axhline(y = 0.5, color = 'r', linestyle = '-')
#### 
plt.plot(x_train,y_train,'o')
plt.plot(x_linear,y_linear,'-r',label='sigmoid')
####       
plt.plot(x, y, '-r', label='sigmoid')
y_pred = logistic.predict(x_train)
print(y_pred)
plt.grid()
plt.show()
    
```
