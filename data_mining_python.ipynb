LINEAR REGRESSION

```python
#اضافه کردن کتابخانه یادگیری ماشین
from sklearn.linear_model import LinearRegression
#اضافه کردن کتابخانه رندوم برای ایجاد اعداد رندوم از روی یک دانه
from random import randint
#اضافه کردن کتابخانه مخصوص رسم نمودار و نقاط
import matplotlib.pyplot as plt
#اضافه  کردن کتابخانه مخصوص پردازش اعداد
import numpy as np
#ایجاد یک شی از روی تابع رگرسیون خطی
linearity=LinearRegression()
#در اینجا اندازه داده آموزش 20 میباشد
#ساخت داده باری تعلیم مدل
train_x=[]
train_y=[]
for i in range(20):
   train_x.append([randint(0,20)])
   train_y.append([randint(0,20)])
#ترسیم نقاط مربوط به داده ای تعلیمی 
plt.plot(train_x,train_y,'ro')
# کمینه سازی اخلاف مربعات غیر  منفی
model=linearity.fit(train_x,train_y)
#یافتن ضریب  برای رسم نقاط
r_sq = model.score(train_x,train_y)
x = np. linspace(0,20,100)
y = float(model.coef_) * x + float(model.intercept_)
plt. plot(x, y, '-r', label='my linear line')
print('coefficient of determination:', r_sq)

#نمایش عرض از مبدا
print('intercept:', model.intercept_)

# نمایش شیب مربوط به معادله خطی یک متغییره
print('slope:', model.coef_) 

# پیش بینی از روی مدل ها (مرحله آخر)
y_pred = model.predict(train_x)
print('Predicted response:', y_pred, sep='\n')


```

Logistic Regression

```python
##اضافه کردن کتابخانه های معروف
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import math

###داده های آموزشی
x_train=[]
y_train=[]
##ساخت داده آموزشی
for i in range(-10,10,1):
     x_train.append([i])
     y_train.append(randint(0,1))
 ##مدل لجستیکی    
logistic=LogisticRegression(solver='liblinear', C=10.0, random_state=0)
##یادگیری مدل لجستیکی
logistic.fit(x_train,y_train)
##رسم تابع سیگموید
x=np.linspace(-20,20,100)
y=1/(1+np.exp(-x))
##محدود کردن برد تابع
plt.ylim(top=1)  
plt.ylim(bottom=0)
##رسم تابع خطی
x_linear=np.linspace(-20,20,100)
y_linear=float(logistic.coef_)*x+float(logistic.intercept_)
##تابع ثابت Y=1/2
plt.axhline(y = 0.5, color = 'r', linestyle = '-')
####رسم توابع
plt.plot(x_train,y_train,'o')
plt.plot(x_linear,y_linear,'-r',label='sigmoid')
####پیش بینی نمونه دو دسته بندی تک متغییره
plt.plot(x, y, '-r', label='sigmoid')
y_pred = logistic.predict(x_train)
print(y_pred)
plt.grid()
plt.show()
    
```

Decision Trees

```python
#اضافه کردن تابع طبقه بندی 
from sklearn.tree import DecisionTreeClassifier
#اضافه کردن کتابخانه مربوط به ترسیم درخت
from sklearn import tree
#اضافه کردن کتابخانه تولید اعداد تصادفی
from random import randint
#ساخت یک شی از روی تابع درخت تصمیم برای طبقه بندی 
dt=DecisionTreeClassifier()
#مجموعه داده ها
x_train=[]
y_train=[]
#تولید اعداد تصادفی و رندوم  و الحاق آن به آخر لیست  
for i in range(5):
    x_train.append([randint(0,100)])
    y_train.append([randint(0,100)])
#چاپ اعداد تصادفی
print(f'x_train points:{x_train}')
print(f'y_train points:{y_train}')
#تابع برازش و کلاس بندی درخت تصمیم
dt=dt.fit(x_train,y_train)
#رسم درخت تصمیم به همراه داده های آن
tree.plot_tree(dt)
```
